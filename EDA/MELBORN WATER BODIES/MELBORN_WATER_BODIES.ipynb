```python
### 1 Bar Graph of Total Capacity vs. Current Volume: Compare the total capacity and current volume for each water storage area.
df_1 = data_load[['Water storage area','Current volume (ML)*','Total capacity (ML)']]
df_1
```

### 2 • Scatter Plot of Current Volume vs. Capacity Change: Analyze the relationship between current volume and capacity changes.
#

```python
#Index(['Water storage area', 'Current volume (ML)*', 'Total capacity (ML)'], dtype='object')
for i in df_1.index.unique():
    print (i)
    df_melted = df_1.loc[i]
    df_melted.set_index('Water storage area', inplace=True)
    df_melted.plot(kind='bar', edgecolor = 'black')
    plt.title(f'Water Storage Data for {i}')
    plt.show()
    print(line_break)
```

# What is the distribution of movie ratings?


```python
rating_counts = data_load['Rating'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
rating_counts.plot(kind='line', marker = '*')
plt.title('Rating Value Counts')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.grid()
plt.show()
```

# How does the number of votes correlate with movie ratings?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named data_load
# Convert relevant columns to numeric if necessary
data_load['Rating'] = pd.to_numeric(data_load['Rating'], errors='coerce')
data_load['No of Persons Voted'] = pd.to_numeric(data_load['No of Persons Voted'], errors='coerce')

# Drop any rows with NaN values in these columns
data_filtered = data_load[['Rating', 'No of Persons Voted']].dropna()

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_filtered, x='No of Persons Voted', y='Rating')
plt.title('Scatter Plot of Votes vs Ratings')
plt.xlabel('Number of Persons Voted')
plt.ylabel('Movie Rating')
plt.show()

# Calculate correlation
correlation = data_filtered['Rating'].corr(data_filtered['No of Persons Voted'])
print(f'Correlation between votes and ratings: {correlation}')
```

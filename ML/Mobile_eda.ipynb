```python
#touch_screen     

data_load['touch_screen'].value_counts().plot(kind='bar')

plt.title('Count of Touch Screen Devices')
plt.xlabel('Touch Screen (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks(rotation=0) 
plt.grid()
plt.show()
```

```python
data_load['four_g'].value_counts().plot(kind='bar')


plt.title('Count of 4G Support Devices')
plt.xlabel('4G Support (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks(rotation=0) 
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data_load.boxplot(column='price_range', by='four_g', grid=False)

# Adding titles and labels
plt.title('Price Range Distribution by 4G Support')
plt.suptitle('')  # Suppress the default title to keep it clean
plt.xlabel('4G Support (0 = No, 1 = Yes)')
plt.ylabel('Price Range')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming data_load is your DataFrame
plt.figure(figsize=(10, 6))

plt.scatter(data_load['px_height'], data_load['px_width'], alpha=0.7)

plt.title('Scatter Plot of Pixel Height vs. Pixel Width')
plt.xlabel('Pixel Height')
plt.ylabel('Pixel Width')
plt.grid(True)
plt.show()
```
